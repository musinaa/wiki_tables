EXTRACTING TABLES IN HTML

Set up

```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site ='https://en.wikipedia.org/wiki/List_of_African_countries_by_GDP_(PPP)'
base_site
```

```python
r =requests.get(base_site)
```

```python
html=r.content
html
```

Extracting tables with Beautiful soup

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
table=soup.find_all('table')[0]
```

```python
table.find_all('tr')[0].contents
```

```python
country= [row.contents[5]text for row in table.find_all('tr')]
country
```

Using Pandas to extract tables
